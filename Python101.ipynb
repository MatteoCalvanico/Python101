```python
#Dizionario

#Memoria associativa (attributo:valore) | "{'Attributo chiave':'valore', 'es':'val'}"

#Ogni chiave è immutabile, quindi possiamo usare tutti i tipi immutabili per rappresentarla
#Ogni chaive può essere ripetuta, ma essa assume il valore dell’ultima occorrenza della stessa

dict = {}                                          # {}

dict['one'] = "key is one"                         # {'one':'key one'}

dict[3] = "key is 3"                               # {'one':'key one',3:'key is 3'}

dict2 = {'name': 'pippo','code':1, 'role': 'dog'}


print (dict['one'])                                # value for key 'one'

print (dict[3])                                    # value for key 3

print (dict2)                                      # complete dictionary

print (dict2.keys())                               # all the keys

print (dict2.values())                             # all the values


#Operatori

del(dict[2])                                       # Cancella l'associazione chiave:valore

list(dict)                                         # Ritorna una lista con tutte la chiavi usate nel dizionario

sorted(dict)                                       # Come il precedente ma in ordine
```

```python
#Controlli e cicli

#if
"Finisco l'if con : e può contenere uno o + elif (aka switch o else if)"

if(x in y):
  #comandi
   print('Fa qualcosa se cond1 TRUE')
elif y == x: #else if
  #comandi
  print('Fa qualcosa se cond1 FALSE e cond2 TRUE')
else:
  #comandi
  print('Fa qualcosa se ne cond1 ne cond2 TRUE')

'------------------------------------------------------------'

#for
"Finisco sempre con :, ed ha una sinstassi tipo foreach"
numbers = [1,2,3,4,5]

for num in numbers:
    print(num)

"Il loop viene fatto attraverso una sequenza, per fare un for classico:"
for i in range(len(numbers)):
  print(i)

'------------------------------------------------------------'

#while
while(x == y):
  print("Faccio questo finchè è TRUE")

'------------------------------------------------------------'

#try... finally, with
"Si usa per i controllo errori, se c'è un errore non si blocca ma fa altro"

try:
  print("do something")
except:
  print("if 'try' didn't work")
finally:
  print("handle all exceptions")
```

```python
#File management

#Leggere                 #'r' indica lettura (NON creiamo il file se non c'è) r+ non creiamo e non cancelliamo se esiste il file, 'w' scrittura (creiamo il file se non c'è) w+ cancelliamo il file se esiste e lo creiamo se non esiste, 'a' append (apriamo il file dal fondo) a+ creiamo il file se non esiste e se esiste non cancelliamo il contenuto
with open("nomefile.txt", "r") as fin:   
    text = fin.readlines() #ci sono altri metodi come .read([count]) se non metto count legge fino alla fine 
    for line in text:
        word = line.split()
        print(word)

#Sul handle del file (fin) possiamo chiamare altre propietà
fin = open("nomefile.txt", "r")

fin.closed      #Torna true se il file è chiuso

fin.mode        #Indica la modalità di apertura del file

fin.name        #Indica il nome fisico del file

fin.seek(0,0)   #Spostiamo l'indice del file di quanto detto dal primo numero, il secondo indica da che posizione iniziare a spostarsi (0 inizio, 1 pos corrente, 2 fine del file)

fin.truncate()  #Eliminiami il contenuto del file (solo se aperto in r+)

fin.tell()      #Posizione del file


#Scrivere
file = open('file.txt','w')
for elem in ['one','two','three']:
  file.write(elem)  #Scrive nel file
file.close()    #Chiude il file
```
